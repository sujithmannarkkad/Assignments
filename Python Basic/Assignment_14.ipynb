```python
1. What does RGBA stand for?

What does RGB and RGBA stand for?
RGB is a 3-channel format containing data for Red, Green, and Blue. 
RGBA is a 4-channel format containing data for Red, Green, Blue, and Alpha. 
There is no use to the Alpha channel other than making the color transparent/opaque (or partially transparent; translucent).
```

```python
2. From the Pillow module, how do you get the RGBA value of any images?

getpixel() witll give you (red, green, blue, alpha) tuples.

from PIL import Image
imgobj = Image.open('profile pic.jpg')
pixels = imgobj.convert('RGBA')
lofpixels = []
w = 1920
h = 1080
for i in range(w):
    for j in range(h):
        r,g,b,a = pixels.getpixel((i,j))
        print(r,g,b,a)
```

```python
3. What is a box tuple, and how does it work?

Many of Pillowâ€™s functions and methods take a box tuple argument. This means Pillow is expecting a tuple of four integer coordinates that represent a rectangular region in an image. The four integers are, in order, as follows:

Left The x-coordinate of the leftmost edge of the box.

Top The y-coordinate of the top edge of the box.

Right The x-coordinate of one pixel to the right of the rightmost edge of the box. This integer must be greater than the left integer.

Bottom The y-coordinate of one pixel lower than the bottom edge of the box. This integer must be greater than the top integer.
```

```python
6. After making changes to an Image object, how could you save it as an image file?

we can use save function to save the new file.

from PIL import Image
image = Image.open('flower.jpg')
new_image = image.resize((400, 400))
new_image.save('flower_400.jpg')

print(image.size) # Output: (1920, 1280)
print(new_image.size) # Output: (400, 400)
```

```python
8. Image objects do not have drawing methods. What kind of object does? How do you get this kind of object?


If you need to draw lines, rectangles, circles, or other simple shapes on an image, use Pillowâ€™s ImageDraw module

from PIL import Image, ImageDraw
im = Image.new('RGBA', (200, 200), 'white')
draw = ImageDraw.Draw(im)
First, we import Image and ImageDraw. Then we create a new image, in this case, a 200Ã—200 white image, 
and store the Image object in im. 
We pass the Image object to the ImageDraw.Draw() function to receive an ImageDraw object. 
This object has several methods for drawing shapes and text onto an Image object. 
Store the ImageDraw object in a variable like draw so you can use it easily in the following example.

draw.line([(0, 0), (199, 0), (199, 199), (0, 199), (0, 0)], fill='black')
draw.rectangle((20, 30, 60, 60), fill='blue')
im.save('drawing.png')
```
